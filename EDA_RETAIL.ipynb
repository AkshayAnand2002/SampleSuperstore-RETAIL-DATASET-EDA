```python
sales_by_category = dataset.groupby('Category').size()
plot_by_category = sales_by_category.plot(title='Sales by Category', color='red')
plot_by_category.set_xlabel('Category')
plot_by_category.set_ylabel('Sales')
```

```python
sales_by_shipmode = dataset.groupby('Ship Mode').size()
sales_by_shipmode = sales_by_shipmode.plot(title='Sales by Ship Mode',color='yellow')
sales_by_shipmode.set_xlabel('Ship Mode')
sales_by_shipmode.set_ylabel('Sales')
```

```python
sales_by_segment = dataset.groupby('Segment').size()
sales_by_segment = sales_by_segment.plot(title='Sales by Segment', color='yellowgreen')
sales_by_segment.set_xlabel('Segment')
sales_by_segment.set_ylabel('Sales')
```

```python
#Contribution of each Sub-category to its total count.
plt.figure(figsize=(10,8))
dataset['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
df_segment= dataset.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(20,22),
                    startangle=90,     
                    shadow=True,
                    labels = df_segment.index)
plt.title('Segment Wise Analysis')
```

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(12,8));
sns.countplot(dataset['Category'],ax=axs[0][0])
sns.countplot(dataset['Segment'],ax=axs[0][1])
sns.countplot(dataset['Ship Mode'],ax=axs[1][0])
sns.countplot(dataset['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=12)
axs[0][1].set_title('Segment',fontsize=12)
axs[1][0].set_title('Ship Mode',fontsize=12)
axs[1][1].set_title('Region',fontsize=12)
plt.tight_layout()
#countplot of each column
```
